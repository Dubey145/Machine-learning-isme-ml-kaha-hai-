```python
#splitting data into 2 parts
#try to pickup 20% of your test data randomly
#sklearn provides a function to split data
from sklearn import model_selection 

print(type(x),type(y))

x_train,x_test,y_train,y_test = model_selection.train_test_split(x,y) #passing numpy arrays
print(x_train.shape)
print(x_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
algo1 = LinearRegression()
type(algo1)
```

```python
algo1.fit(x_train,y_train)
```

```python
y_pred=algo1.predict(x_test) #these are predictions 
#compare results of y_pred with y_test
#can be done by plotting
import matplotlib.pyplot as plt
plt.scatter(y_pred,y_test)
plt.axis([0,40,0,40])
x = [i for i in range (0,51)]
y=x
plt.plot(x,y)
'''
x_axis has the predictions whereas y axis has actual output
if a point lies on the line y = x it has been accurately predicted, 
if above then its actual value was more than expected
if below then the actual value was less then expected
'''
```
